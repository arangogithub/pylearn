# Nonlinear transport equation

*  $ u_t + u\, u_x=0$
* Initial condition $u(0,x) =f(x)$
* Implicit solution given by $u = f(x-t\,u),$ where $u=u(t,x)$ or $t$ and $x$ given

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
pi = np.pi

def f(x): return pi/2 - np.arctan(x)

def F(t, x,u): return  u - f(x - u*t)
```

```python
tt = np.linspace(0.0,5.0,600)
xx = np.linspace(-2.0.0,7.0,600)
uu = np.linspace(-0,pi,600)
```

```python

X, U = np.meshgrid(xx, uu)
plt.figure(figsize=(10,7))
plt.subplot(3, 1, 1)
t = 0.0
levels = [ 0.0]
plt.title(r'$t=0$')
plt.yticks([0, pi/2, pi],[r'$0$', r'$\pi/2$', r'$\pi$'])
plt.contour(X, U, F(t, X, U), levels, linewidths = 2.5, colors="black" )

plt.subplot(3, 1,2)
t = 0.5
levels = [ 0.0]
plt.title(r'$t=0.5$')
plt.yticks([0, pi/2, pi],[r'$0$', r'$\pi/2$', r'$\pi$'])
plt.contour(X, U, F(t, X, U), levels, linewidths = 2.5, colors="black" )

plt.subplot(3, 1,3)
t = 1.0
levels = [ 0.0]
plt.title(r'$t=1.0$')
plt.yticks([0, pi/2, pi],[r'$0$', r'$\pi/2$', r'$\pi$'])
plt.contour(X, U, F(t, X, U), levels,  linewidths = 2.5, colors="black" )

plt.tight_layout()
plt.savefig('noshock_still.pdf', bbox_inches='tight')
```

# Multive value region

```python
tt = np.linspace(1.0,5.0,100)

xx_u = tt*f(np.sqrt(tt-1))+np.sqrt(tt-1)
xx_l = tt*f(-np.sqrt(tt-1))-np.sqrt(tt-1)
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1, 1, 1)
plt.title('multivalue region', fontsize=25)
plt.xlabel(r'$t$', fontsize=25)
plt.ylabel(r'$x$', fontsize=25)
plt.plot(tt,xx_u, linewidth = 2.5, color="green", label=r'$x_u$')
plt.plot(tt,xx_l,linewidth = 2.5, color="blue", label=r'$x_l$' )
plt.legend(loc='upper left', fontsize=25)
plt.savefig('multivalue_region.pdf', bbox_inches='tight')
```

```python
def x_u(t): return t*f(-np.sqrt(t-1)) - np. sqrt(t-1)
def x_l(t): return t*f(np.sqrt(t-1)) + np. sqrt(t-1)
```

```python

X, U = np.meshgrid(xx, uu)
plt.figure(figsize=(10,7))

plt.subplot(3, 1, 1)
t = 1.5
plt.vlines(x_l(t), 0, pi, colors='green', linewidths = 2.5)
plt.vlines(x_u(t), 0, pi, colors='blue', linewidths = 2.5)
levels = [ 0.0]
plt.title(r'$t=1.5$')
plt.yticks([0, pi/2, pi],[r'$0$', r'$\pi/2$', r'$\pi$'])
plt.contour(X, U, F(t, X, U), levels, linewidths = 2.5, colors="red")

plt.subplot(3, 1,2)
t = 2.5
plt.vlines(x_l(t), 0, pi, colors='green', linewidths = 2.5)
plt.vlines(x_u(t), 0, pi, colors='blue', linewidths = 2.5)
levels = [ 0.0]
plt.title(r'$t=2.5$')
plt.yticks([0, pi/2, pi],[r'$0$', r'$\pi/2$', r'$\pi$'])
plt.contour(X, U, F(t, X, U), levels, linewidths = 2.5, colors="red")

plt.subplot(3, 1,3)
t = 3.0
plt.vlines(x_l(t), 0, pi, colors='green', linewidths = 2.5)
plt.vlines(x_u(t), 0, pi, colors='blue', linewidths = 2.5)
levels = [ 0.0]
plt.title(r'$t=3.0$')
plt.yticks([0, pi/2, pi],[r'$0$', r'$\pi/2$', r'$\pi$'])
plt.contour(X, U, F(t, X, U), levels, linewidths = 2.5, colors="red")

plt.tight_layout()
plt.savefig('multivalue_sol.pdf', bbox_inches='tight')
```
