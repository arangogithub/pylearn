```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vidaH = 114
vidaM = 23
iH = [1]+[0 for j in range(vidaH-1)]
iM = [0 for j in range(vidaM)]
```

```python
def proxEstadoH(eh):
    nuevasH = eh[32]+ eh[50]+ eh[68]+ eh[86]+ eh[104] 
    ah = [ eh[j] for j in range(vidaH-1)]
    return [nuevasH] + ah
```

```python
def proxEstadoM(eh,em):
    nuevosM = eh[41]+ eh[59]+ eh[77]+ eh[95]+ eh[113] 
    am = [ em[j] for j in range(vidaM-1)]
    return [nuevosM] + am 
```

```python
def evolH(numMeses, eh):
    contador = 1
    ah = eh
    while contador < numMeses + 1:
        ah = proxEstadoH(ah)
        contador = contador + 1
    return ah
```

```python
def evol(numMeses,eh,em):
    contador = 1
    ah,am = eh, em 
    while contador < numMeses + 1:
        am = proxEstadoM(ah,am)
        ah = proxEstadoH(ah)
        contador = contador + 1
    return (ah,am)
```

```python
z = evolH(12*10, iH)
sum(z)
```

```python
plt.plot(z)
plt.show()
```

```python
hatoH = [sum(evolH(k, iH)) for k in range(12*15)]
```

```python
plt.plot(hatoH)
plt.show()
```

```python
def venta(n):
    z = evol(n, iH, iM)
    return z[1][-1]
```

```python
evolVenta = [venta(j) for j in range(12*18)]
plt.plot(evolVenta)
plt.show()
```

```python
fig = plt.figure(figsize=(30,20))
ax = fig.add_subplot(111)
ax.plot(evolVenta, lw = 7, label='venta')
ax.plot(hatoH, lw = 7, label='hato')
legend = ax.legend(loc='upper right', shadow=True, fontsize='x-large')
legend.get_frame().set_facecolor('#00FFCC')
plt.show()
```
