```python
#matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy import init_printing
from sympy import *
```

```python
x, y, t = symbols('x y t')
exprs_u = -x-Rational(1,2)*(y**2) - 3*x**4 + y*x**3 +18*y**2*x**2 -x*y**3 -3*y**4 + Rational(1,500)
u = lambdify((x, y), exprs_u, "numpy") 
```

```python
detm = simplify(diff(exprs_u,x,x)*diff(exprs_u,y,y) - diff(exprs_u,x,y)**2)
laplacian = simplify(diff(exprs_u,x,x)+diff(exprs_u,y,y))
jacD = lambdify((x, y), detm, "numpy") 
```

```python
exprs_du_theta = diff(exprs_u,x)*sp.cos(t) + diff(exprs_u,y)*sp.sin(t)
du_t = lambdify((t,x, y), exprs_du_theta, "numpy")  
```

```python
expr_G = diff(exprs_u,x,x)- 1j* diff(exprs_u,x,y)
G_funct  = lambdify((x, y), expr_G, "numpy") 
 
```

```python
curv =  -(diff(exprs_u,x,x)*diff(exprs_u,y)**2 + diff(exprs_u,y,y)*diff(exprs_u,x)**2  -
         2*diff(exprs_u,x,y)*diff(exprs_u,x)*diff(exprs_u,y))

kappa =   lambdify((x, y), curv, "numpy")
```

```python
wd = 0.25
N = 1000
xx, yy  = np.linspace(-wd,wd/10,N), np.linspace(-wd,wd,N)
X, Y = np.meshgrid(xx, yy)
Jdet = jacD(X,Y)
U = u(X,Y)
DU1 = du_t(np.pi/2,X,Y)
DU2 = du_t(1.003*np.pi/2,X,Y)
DU3 = du_t(1.007*np.pi/2,X,Y)
K = kappa(X,Y)
```
